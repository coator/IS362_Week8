```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_fwf("auto-mpg.data")
df.columns=('mpg','cylinders','displacement','horsepower','weight','acceleration','model year','origin','car name')
```

```python
df.replace('?',np.nan,inplace=True)
df['horsepower'] = df['horsepower'].astype(float)
df.dtypes
```

```python
df['origin'].replace(1,'USA',inplace=True)
df['origin'].replace(2,'Asia',inplace=True)
df['origin'].replace(3,'Europe',inplace=True)
```

```python
df['cylinders'].hist()
```

```python
df.plot.scatter('horsepower','weight',figsize=(5,5))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
usa_cars = df['origin'] == 'USA'
asia_cars = df['origin'] == 'Asia'
europe_cars = df['origin'] == 'Europe'
```

#### Question: How did mpg change over time?

```python
fig, ax = plt.subplots(figsize=(10,10))
uscar =ax.scatter(df.loc[usa_cars]['model year'],df.loc[usa_cars]['mpg'])
uscar.set_color('orange')

asiacar = ax.scatter(df.loc[asia_cars]['model year'],df.loc[asia_cars]['mpg'])
uscar.set_color('red')

europecar = ax.scatter(df.loc[europe_cars]['model year'],df.loc[europe_cars]['mpg'])
europecar.set_color('blue')

ax.set(xlabel='Year', ylabel='MPG', title="MPG over the years per manufacturer")
```

#### Question: How did weight change on cars over time by origin?

```python
sns.set_theme()

g = sns.lmplot(
    data=df,
    x="model year", y="weight", hue="origin",
    height=10
)
```
